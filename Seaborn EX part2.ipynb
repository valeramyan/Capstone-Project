# MatrixPlot





--------

```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.corr()
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='inferno')
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights.head()
```

```python
flights.pivot_table(index='year', columns='month',values='passengers')
```

```python
flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp= flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,lw=1)
```

```python
sns.heatmap(fp,lw=1,cmap='coolwarm')
```

```python
sns.clustermap(fp,lw=2)
```

```python
sns.clustermap(fp,lw=2,standard_scale=2)
```

# Grids
-------

```python
iris= sns.load_dataset('iris')
import matplotlib.pyplot as plt
```

```python
iris.head()
```

```python
sns.pairplot(iris)
sns.set_style('darkgrid')
```

```python
sns.PairGrid(iris)
```

```python
g= sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g= sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(plt.scatter)
```
