```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
 df = pd.read_csv('C:\\Users\\flang\\Desktop\\python\\DATAFRAME\\Movie-Ratings.csv')
    
```

```python
df.head()
df.columns=['Film', 'Genre', 'CriticRatings', 'AudienceRatings','Budgetmillion','Year']
df.Film= df.Film.astype('category')
df.Genre= df.Genre.astype('category')
df.Year=df.Year.astype('category')
```

```python

f,axes=plt.subplots(2,2,figsize=(12,6),sharex=True,sharey=True)
k1=sns.kdeplot(df.Budgetmillion, df.AudienceRatings,ax=axes[0,0])
k2=sns.kdeplot(df.Budgetmillion, df.CriticRatings,ax=axes[0,1] )
k3= sns.boxplot(data= df[df.Genre == 'Drama'],x='Year', y='CriticRatings',ax=axes[1,1])
k4=sns.kdeplot(df.CriticRatings,df.AudienceRatings, shade=True,shade_lowest=False, cmap='Reds',ax=axes[1,0] )
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
f,axes=plt.subplots(2,2,figsize=(15,15),sharex=True,sharey=True)
k1=sns.kdeplot(df.Budgetmillion, df.AudienceRatings,ax=axes[1,1])
k2=sns.kdeplot(df.Budgetmillion, df.CriticRatings,ax=axes[0,1] )
k3=sns.kdeplot(df.Budgetmillion, df.AudienceRatings,shade = True,shade_lowest=True,cmap='inferno',ax=axes[1,0])
k3b=sns.kdeplot(df.Budgetmillion,df.AudienceRatings,cmap='cool',ax=axes[1,0])
k4=sns.kdeplot(df.Budgetmillion,df.AudienceRatings, shade=True,shade_lowest=True, cmap='inferno',ax=axes[0,0] )
k4b=sns.kdeplot(df.Budgetmillion,df.AudienceRatings, cmap='rainbow',ax=axes[0,0] )
k1.set(xlim=(-20,160))
```
