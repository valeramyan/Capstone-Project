# legend

```python
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label= 'x squared')
ax.plot(x,x**3,label='x cube')
ax.legend(loc=0)
```

# Coloring dash

```python
fig= plt.figure(figsize=(6,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, color='orange', lw= 10, ls= '--',marker='s',ms=10, markeredgecolor='black')
```

```python
fig= plt.figure(figsize=(8,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, color='orange', lw= 3, ls='-.')
```

```python
fig= plt.figure(figsize=(8,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, color='orange', lw= 2, ls='steps',marker='s', ms=10)
```

```python
fig= plt.figure(figsize=(8,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, color='orange', lw= 2, ls='steps',marker='s', ms=10, markerfacecolor='black', markeredgewidth=4)
plt.show()
```
