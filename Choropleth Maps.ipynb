# Choropleth Maps 

```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:\\Users\\flang\\Downloads\\2014_World_Power_Consumption')
```

```python
data= dict(type='choropleth',
          locations= df1['Country'],
          locationmode='country names',
          z= df1['Power Consumption KWH'],
          text=df1['Country'],
          colorbar= {'title':'2014 Consumption'})
```

```python
layout= dict(title = '2014 World PowerConsumption',
            geo=dict(showframe=False,
                    projection= {'type':'Mercator'}))
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```

```python
df2 = pd.read_csv('C:\\Users\\flang\\Downloads\\2012_Election_Data')
```

```python
data= dict(type= 'choropleth',
          colorscale = 'Viridis',
           reversescale=True,
          locations= df2['State Abv'],
          z= df2['Voting-Age Population (VAP)'],
          locationmode= 'USA-states',
          text=df2['State'],
          marker = dict(line = dict(color = 'rgb(255,255,255)',width = 2)),
          colorbar = {'title':"Voting-Age Population"}
            ) 
```
